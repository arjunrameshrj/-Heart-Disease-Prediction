# Introduction to the Heart Disease Prediction Dataset:

This dataset is designed to facilitate the analysis and prediction of heart disease based on several key health attributes. The dataset includes information about individuals, such as their age, gender, and various physiological measurements. The objective is to use these features to predict whether an individual is likely to have heart disease or not.

* Objective of the study

The primary objective of the heart disease prediction dataset is to develop a predictive model that can accurately identify the likelihood of individuals having heart disease based on their health-related features. Specifically:

# Features:
* Age: The age of the individual, providing insight into age-related factors.
* Sex: Gender of the individual, allowing for gender-based analysis.
* Chest Pain (cp): Categorization of chest pain types, a crucial symptom of heart-related issues.
* Resting Blood Pressure (trtbps): The blood pressure level at rest, indicating cardiovascular health.
* Cholesterol (chol): Serum cholesterol level, a key factor in heart health assessment.
* Fasting Blood Sugar (fbs): Binary indication of fasting blood sugar level (> 120 mg/dl).
* Resting Electrocardiographic Results (restecg): Categorical data reflecting electrocardiogram findings at rest.
* Maximum Heart Rate (thalachh): The highest heart rate achieved during exercise, relevant to cardiovascular fitness.
* Exercise-Induced Angina (exng): Binary indication of angina induced by exercise.
* ST Depression (oldpeak): ST segment depression induced by exercise relative to rest.
* Slope of Peak Exercise ST Segment (slp): Describes the slope of the ST segment during peak exercise.
* Number of Colored Vessels (caa): Count of major vessels colored by fluoroscopy, providing insight into vascular health.
* Thalassemia (thall): Categorical data related to thalassemia, a blood disorder.
* Output: Binary classification indicating the presence (1) or absence (0) of heart disease.

```python
X = df.drop('output', axis=1)
y = df['output']
```

 # Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert the scaled features back to a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)


```

# Splitting Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

# Model Training
 # MODEL CREATION
 Classification Algorithms
 1. K-Nearest Neighbor (KNN)
 2. Naive-Bayes
 3.Support Vector Machine
 4.Decision Tree
 5.Random Forest
 6.Ada Boost
 7. XG Boost
 8. Logistic regression








```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression


models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes (Gaussian)": GaussianNB(),
    "Naive Bayes (Bernoulli)": BernoulliNB(),
    "Support Vector Machine": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Ada Boost": AdaBoostClassifier(n_estimators=10, learning_rate=1.0),
    "XG Boost": XGBClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    print()
```

 # accuracy score comparison

```python


comparison = pd.DataFrame({'Model': ['K-Nearest neighbor', 'Naive BayesGNB', 'Naive Bayes-BNB', 'Support Vector Machine', 'Decision Tree', 'Random Forest', 'Ada Boost', 'XG Boost', 'Logistic Regression'],
                            'Accuracy': [68.13,84.62,81.32,71.43,81.32,84.62,76.92,82.42,82.42],})


plt.figure(figsize=(18, 5))
sns.barplot(x='Model', y='Accuracy',color='y',data=comparison)

# Annotate each bar with the percentage value
for index, value in enumerate(comparison['Accuracy']):
    plt.text(index, value + 1, f'{value:.2f}%', ha='center', va='bottom')


plt.show()
```

# **Observation**

* The most accurate model is Nive BayesGNB,and Random Forest with an accuracy of 84.62%.
* Based on the classification reports for each model, the best model overall is **Random Forest**. It has the highest accuracy (84.62%), as well as a good balance of precision, recall, and F1-score for both classes.
* The least accurate model is K-Nearest Neighbors, with an accuracy of 68.13%.

 Oversampling is a technique used in machine learning to address class imbalance by increasing
the number of instances in the minority class (the less frequent class). This helps to balance the
class distribution, which can lead to better model performance

# SMOTE -for data balancing-synthetic minority over sampling Technique
